```python
import pandas as pd
from IPython.display import display

def read_data_frame(document_id, sheet_name):
    export_link = f"https://docs.google.com/spreadsheets/d/{document_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"
    return pd.read_csv(export_link)

document_id = '14fKHsblfqZfWj3iAaM2oA51TlYfQlFT4WKo52fVaQ9U'
products_df = read_data_frame(document_id, 'products')
emails_df = read_data_frame(document_id, 'emails')

```

```python
def classify_email_with_openai(subject, message):
    # Combine subject and message into one prompt for context
    prompt = f"The subject of the email is: '{subject}'. The message is: '{message}'.\n\nBe concise and respond with only the classification: product inquiry, order request, or unknown. Do not include any quotation marks."

    # Call the OpenAI API to classify the email
    response = client.chat.completions.create(
        model="gpt-4o",  # Use the provided model
        messages=[
            {"role": "user", "content": prompt}
        ]
    )

    # Extract the classification and remove any surrounding quotes
    classification = response.choices[0].message.content.strip().replace("'", "").replace('"', "").lower()

    return classification

# Apply this function to the emails DataFrame
emails_df['category'] = emails_df.apply(lambda row: classify_email_with_openai(row['subject'], row['message']), axis=1)

# Display the updated DataFrame with classifications
display(emails_df[['email_id', 'subject', 'category']])
```

```python
def generate_order_response(email_id, order_status_df, products_df):
    """
    Generate a personalized response email based on the order status.
    """
    order_details = order_status_df[order_status_df['email_id'] == email_id]
    response_lines = []

    for _, order in order_details.iterrows():
        product_row = products_df[products_df['product_id'] == order['product_id']]
        product_name = product_row['name'].values[0] if not product_row.empty else "Unknown Product"

        if order['status'] == 'created':
            response_lines.append(f"Your order for {order['quantity']} x {product_name} has been successfully processed.")
        elif order['status'] == 'out of stock':
            response_lines.append(f"Unfortunately, the product {product_name} is currently out of stock, and we could only fulfill {order['quantity']} units.")
        elif order['status'] == 'product not found':
            response_lines.append(f"We could not find the product {order['product_id']} in our catalog.")

    if response_lines:
        response_message = "Dear Customer,\n\n" + "\n".join(response_lines) + "\n\nThank you for your order!\nBest regards,\nCustomer Service Team"
    else:
        response_message = "Dear Customer,\n\nWe could not process your order due to an issue with the products mentioned. Please check the details and try again.\n\nBest regards,\nCustomer Service Team"

    return response_message

# Generate responses for all order requests and populate the order-response sheet
order_requests_df['response'] = order_requests_df['email_id'].apply(lambda email_id: generate_order_response(email_id, order_status_df, products_df))
display(order_requests_df[['email_id', 'response']])

# At this point, you should save the order_requests_df (with responses) to your desired output format (e.g., spreadsheet).
```

```python
def extract_relevant_products(message, products_df):
    """
    Extract product information relevant to the customer's inquiry from the product catalog.
    """
    relevant_products = []

    for _, product_row in products_df.iterrows():
        product_name = product_row['name'].lower()

        # Check if product name is mentioned in the message
        if product_name in message.lower():
            relevant_products.append(product_row)

    return relevant_products

```

```python
def generate_inquiry_response(email_id, message, products_df):
    """
    Generate a response to a product inquiry using relevant information from the product catalog.
    """
    relevant_products = extract_relevant_products(message, products_df)

    if relevant_products:
        response_lines = []

        for product_row in relevant_products:
            product_name = product_row['name']
            product_description = product_row['description']
            product_price = product_row['price']
            stock_status = "in stock" if product_row['stock'] > 0 else "out of stock"

            response_lines.append(
                f"{product_name}: {product_description} Currently priced at ${product_price}. Stock status: {stock_status}."
            )

        response_message = "Dear Customer,\n\nThank you for your inquiry. Here are the details for the products you asked about:\n\n" + "\n\n".join(response_lines) + "\n\nBest regards,\nCustomer Service Team"
    else:
        response_message = "Dear Customer,\n\nThank you for your inquiry. Unfortunately, we could not find the products you mentioned in our catalog. Please check the product names and try again.\n\nBest regards,\nCustomer Service Team"

    return response_message

# Generate responses for all product inquiries and populate the inquiry-response sheet
inquiry_requests_df = emails_df[emails_df['category'] == 'product inquiry']
inquiry_requests_df['response'] = inquiry_requests_df.apply(lambda row: generate_inquiry_response(row['email_id'], row['message'], products_df), axis=1)
display(inquiry_requests_df[['email_id', 'response']])
```
